```python
import nltk
from nltk.stem import *
from nltk.corpus import stopwords 
from nltk.corpus import inaugural
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('inaugural')
nltk.download('stopwords')
```

```python
names=inaugural.fileids()
print(len(names))
names[0],names[55]
```

```python
#Task A

text51=""+inaugural.raw(fileids=names[51])+inaugural.raw(fileids=names[52])+inaugural.raw(fileids=names[53])+inaugural.raw(fileids=names[54])+inaugural.raw(fileids=names[55])
text51=str(text51)
print(text51)
print(" ")

#Task B

stk51=sent_tokenize(text51)
print(stk51)
#print(text51)
```

```python
#Task C

for w in stk51:
    w51=word_tokenize(w)
    print(w51)
```

```python
#Task D

stop_words = set(stopwords.words('english')) 
wtk51=word_tokenize(text51)
stop_51=[]
for w in wtk51:
    if w not in stop_words: 
        stop_51.append(w)
print(stop_51)
print("   ")
ps = PorterStemmer()
stem_51=[]
for z in stop_51:
    z=ps.stem(z)
    stem_51.append(z)
print(stem_51)
```

```python
#Task I


wl=[]
task=[]
stop_words = set(stopwords.words('english')) 
word_freq=word_tokenize(text51)
for w in word_freq:
    if w not in stop_words:
        wl.append(w)
sent_freq=sent_tokenize(text51)
for each in sent_freq:
    line=each.split()
    sumc=0
    for k in line:
        sumc=sumc+wl.count(k)
    l=[]
    l.append(each)
    l.append(sumc)

    task.append(l)
#print(word_freq)
#print(wl)
task.sort(key=lambda x: -x[1])
for i in range(0,5):
    print(task[[i][0]])
```

```python
#Task H

from nltk import pos_tag, word_tokenize
ps=[]
ps=list(pos_tag(word_tokenize(text51)))
#print(ps)
tags=[]
tags=[["CC",0],["CD",0],["DT",0],["EX",0],["FW",0],["IN",0],["JJ",0],["JJR",0],["JJS",0],["LS",0],["MD",0],["NN",0],["NNS",0],["NNP",0],["NNPS",0],["PDT",0],["POS",0],["PRP",0],["PRP$",0],["RB",0],["RBR",0],["RBS",0],["RP",0],["SYM",0],["TO",0],["UH",0],["VB",0],["VBD",0],["VBG",0],["VBN",0],["VBP",0],["VBZ",0],["WDT",0],["WP",0],["WP$",0],["WRB",0]]
ck=[]
for h in tags:
    ck.append(h[0])

for (a,b) in ps:
    if b in ck:
        v=ck.index(b)
        tags[v][1]=tags[v][1]+1
print(tags)

from wordcloud import WordCloud
import matplotlib.pyplot as plt
wc=""
for g in tags:
    for f in range(0,g[1]):
        wc=wc+" "+g[0]
#print(wc)
wordcloud = WordCloud(width=3000, height=3000, max_words=36,collocations=False).generate(wc) 
#plt.figure() 
plt.imshow(wordcloud, interpolation="bilinear") 
plt.axis("off") 
plt.margins(x=0, y=0) 
plt.show()       
    
    
```

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib as plt
import matplotlib.pyplot as plt 
import pandas as pd 
from textblob import TextBlob 

names=inaugural.fileids()
print(len(names))
names[0],names[55]
text51=""+inaugural.raw(fileids=names[51])+inaugural.raw(fileids=names[52])+inaugural.raw(fileids=names[53])+inaugural.raw(fileids=names[54])+inaugural.raw(fileids=names[55])
text51=str(text51)
print(text51)
print(" ")

blobject = TextBlob(text51)
print(blobject.tags)
nouns = ""
adjs = ""
nounparts = ["NN", "NNS", "NNP", "NNPS"]
adjparts = ["JJ","JJR", "JJS"]
for w in blobject.tags:
    if w[1] in nounparts:
        nouns = nouns + w[0] + " "
    if w[1] in adjparts:
        adjs = adjs + w[0] + " "
print(nouns)
print(adjs)
```
