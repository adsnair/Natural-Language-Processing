```python
import nltk
from nltk.stem import *
from nltk.corpus import stopwords 
from nltk.corpus import inaugural
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('inaugural')
nltk.download('stopwords')
```

```python
names=inaugural.fileids()
print(len(names))
names[0],names[55]
```

```python
#Task A

text51=""+inaugural.raw(fileids=names[51])+inaugural.raw(fileids=names[52])+inaugural.raw(fileids=names[53])+inaugural.raw(fileids=names[54])+inaugural.raw(fileids=names[55])
text51=str(text51)
print(text51)
print(" ")

#Task B

stk51=sent_tokenize(text51)
print(stk51)
#print(text51)
```

```python
#Task C

for w in stk51:
    w51=word_tokenize(w)
    print(w51)
```

```python
#Task D

stop_words = set(stopwords.words('english')) 
wtk51=word_tokenize(text51)
stop_51=[]
for w in wtk51:
    if w not in stop_words: 
        stop_51.append(w)
print(stop_51)
print("   ")
ps = PorterStemmer()
stem_51=[]
for z in stop_51:
    z=ps.stem(z)
    stem_51.append(z)
print(stem_51)
```

```python
#Task F

from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
text=""
for word in stem_51:  
        text = text + " " + word 
wordcloud = WordCloud(width=480, height=480, max_words=10).generate(text) 
plt.figure() 
plt.imshow(wordcloud, interpolation="bilinear") 
plt.axis("off") 
plt.margins(x=0, y=0) 
plt.show() 

from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
text=""
for word in wtk51:  
        text = text + " " + word 
wordcloud = WordCloud(width=480, height=480, max_words=10).generate(text) 
plt.figure() 
plt.imshow(wordcloud, interpolation="bilinear") 
plt.axis("off") 
plt.margins(x=0, y=0) 
plt.show() 
```

```python
#Task I


wl=[]
task=[]
stop_words = set(stopwords.words('english')) 
word_freq=word_tokenize(text51)
for w in word_freq:
    if w not in stop_words:
        wl.append(w)
sent_freq=sent_tokenize(text51)
for each in sent_freq:
    line=each.split()
    sumc=0
    for k in line:
        sumc=sumc+wl.count(k)
    l=[]
    l.append(each)
    l.append(sumc)
    task.append(l)
#print(word_freq)
#print(wl)
task.sort(key=lambda x: -x[1])
for i in range(0,5):
    print(task[[i][0]])
```
